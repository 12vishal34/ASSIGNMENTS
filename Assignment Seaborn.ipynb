```python
Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
```

```python
1.Scatterplot: A scatterplot is a plot that displays the relationship between two continuous variables. It is useful for identifying patterns or trends in the data, as well as identifying correlations, clusters, or outliers.

2.Lineplot: A line plot is a plot that displays the relationship between two continuous variables, with a line connecting the points. It is useful for visualizing trends over time or across different groups.

3.Barplot: A bar plot is a plot that displays the distribution of a categorical variable. It is useful for comparing the frequencies of different categories or subgroups.

4.Heatmap: A heatmap is a plot that displays the correlation between two or more variables, with colors representing the strength of the correlation. It is useful for identifying patterns or trends in large datasets.

5.Histogram: A histogram is a plot that displays the distribution of a single continuous variable, with bars representing the frequency or proportion of each value. It is useful for identifying the shape of the data distribution and any outliers or gaps.



```

```python
Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.
```

```python
import seaborn as sns
fmri=sns.load_dataset('fmri')
```

```python
sns.lineplot(x = fmri.timepoint , y = fmri.signal)
```

```python
Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.
```

```python
titanic=sns.load_dataset('titanic')
```

```python
sns.boxplot(x = titanic.pclass, y = titanic.age)
```

```python
sns.boxplot(x = titanic.pclass, y = titanic.fare)
```

```python
Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.
```

```python
diamonds=sns.load_dataset('diamonds')
```

```python
sns.histplot(data=diamonds, x='price', hue='cut')
```

```python
Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.
```

```python
iris=sns.load_dataset('iris')
```

```python
sns.histplot(data=iris, x='species', hue='species')
```

```python
Que 6: Use the "flights" dataset from seaborn to plot a heatmap.
```

```python
flights=sns.load_dataset('flights')
```

```python
flight = flights.pivot('month', 'year', 'passengers')


sns.heatmap(data=flight ,cmap='coolwarm')
```
